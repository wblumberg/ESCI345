```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

$f(v) = \left(\frac{m}{2\pi kT}\right)^{\frac{3}{2}} 4\pi v^2 e^{-\frac{mv^2}{2kT}}$

$m$ - mass of the particle

$T$ - temperature of the gas

$v$ - speed of the particle

```python
plancks_const_h = 6.625e-34 # Js
boltzmanns_constant_k_b = 1.38e-23 # J K-1

def maxwell_boltzmann_dist(temperature, speed, mass):
    # temperature - Kelvin
    # speed - m/s
    # mass - kg
    
    mult1 = np.power(mass/(2. * np.pi * boltzmanns_constant_k_b * temperature), 3./2.)
    mult2 = 4 * np.pi * np.power(speed, 2.)
    mult3 = np.exp(-(mass * np.power(speed,2))/(2. * boltzmanns_constant_k_b * temperature))
    
    return mult1 * mult2 * mult3
```

```python
av_number = 6.0221408e+23 # particles/mol

speeds = np.arange(1,3000,10) # m/s
temperature = 283. # Kelvin
molarmass = 4e-3 # kg mol-1 (for helium 4)
mass = molarmass * (1./av_number) # for one particle 
dist = maxwell_boltzmann_dist(temperature, speeds, mass)
plt.plot(speeds, dist, label=f'T: {temperature} K')
plt.xlabel(r"Speed [m/s]")
plt.ylabel(r"$f(v)$")
plt.legend()
plt.grid()
plt.show()
```
