```python
%matplotlib inline
import miepython
import rayleigh
import matplotlib.pyplot as plt
import numpy as np
```

### Scratch Paper for Coding

```python
x = 10000 # In the Rayleigh scattering regime
m = 1.33-0.00000001j # index of refraction
angles = np.linspace(-180,180,360) # variety of angles
mu = np.cos(np.radians(angles)) # cos(theta)

qe, qs, qb, g = miepython.mie(m, x)
qext = miepython.i_unpolarized(m, x, mu)

print(g)

def heyney_greenstein(g, mu):
    num = (1. - np.power(g, 2))
    den = np.power((1. + np.power(g, 2) - 2. * g * mu), 3./2.)
    return num/den

for g in [0,0.2,0.4,0.6,0.8]:
    p_hg = heyney_greenstein(g=g, mu=mu)

    plt.plot(angles, p_hg, label=g)
plt.legend()
plt.yscale('log')
plt.show()
#print(heyney_greenstein(g=1., mu=mu))
```

```python
x = 7 # In the Rayleigh scattering regime
m = 1.33-0.0000001j # index of refraction
angles = np.linspace(-180,180,360) # variety of angles
mu = np.cos(np.radians(angles)) # cos(theta)

qe, qs, qb, g = miepython.mie(m, x) # Gets the asymmetry parameter (g)
qext = miepython.i_unpolarized(m, x, mu) # Calculates the extinction for a set number of angles (scattering phase function)

p_hg = heyney_greenstein(g=g, mu=mu) # Gets the H-G scattering phase function

plt.plot(angles, p_hg/np.max(p_hg), label='Heyney-Greenstein Function')
plt.plot(angles, qext/np.max(qext), label='True Phase Function')
plt.legend()
plt.yscale('log')
plt.show()
#print(heyney_greenstein(g=1., mu=mu))
```
