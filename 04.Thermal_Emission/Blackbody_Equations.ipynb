```python
%matplotlib inline
import blackbody as bb
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

ureg = bb.pint.get_application_registry()

sns.set_theme()
```

# Comparing the equations

This first plot really demonstrates the issue classical physics had with predicting the blackbody spectra.  The Rayleigh-Jeans law demonstrates the "ultraviolet catastrophy", where classical physics predicted that an infinite amount of UV radiation would be emitted by any object.

```python
blackbody_temperature = 6000. * ureg.Unit("K")
wlen = np.arange(.01,3,.01) * ureg.Unit("micrometers")# to meters

rj_intensity = bb.rayleigh_jeans(wlen, blackbody_temperature).to("W / meter ** 2 / nanometer", "spectroscopy")
wien_intensity = bb.wien_law(wlen, blackbody_temperature).to("W / meter ** 2 / nanometer", "spectroscopy")
p_intensity = bb.planck(wlen, blackbody_temperature).to("W / meter ** 2 / nanometer", "spectroscopy")

plt.figure(figsize=(7,4))
plt.plot(wlen.to("nanometer", "spectroscopy"), rj_intensity, label="Rayleigh-Jeans Law")
plt.plot(wlen.to("nanometer", "spectroscopy"), wien_intensity, label="Wien's Law")
plt.plot(wlen.to("nanometer", "spectroscopy"), p_intensity, label="Planck's Law")
#plt.yscale('log')
plt.ylim(0,5e13 * 1e-9)
plt.xlabel("Wavelength [$nm$]")
plt.ylabel("Spectral Radiance [$W\ m^{-2}\ sr^{-1}\ nm^{-1}$]")
plt.legend(loc=0)
plt.title("Spectral Radiance at $T$ = 6000 K")
```

# Emission Spectra from the Earth and Sun

```python
wlen = np.arange(.01,100,.01) * ureg.Unit("micrometers")# to meters

sun_temperature = 6000. * ureg.Unit("K")
earth_temperature = 283 * ureg.Unit("K")

p_intensity_sun = bb.planck(wlen, sun_temperature).to("W / meter ** 2 / nanometer", "spectroscopy")
p_intensity_earth = bb.planck(wlen, earth_temperature).to("W / meter ** 2 / nanometer", "spectroscopy")

wlen = wlen.to("nanometer", "spectroscopy")

plt.figure(figsize=(7,4))
plt.plot(wlen, p_intensity_sun, label="$T_{sun}$ = 6000 Kelvin")
plt.plot(wlen, p_intensity_earth, label="$T_{Earth}$ = 293 Kelvin")

plt.yscale('log')
plt.xscale('log')
plt.ylim(1e-3,5e13 * 1e-9)
plt.xlabel("Wavelength [$nm$]")
plt.ylabel("Spectral Radiance [$W\ m^{-2}\ nm^{-1}$]")
plt.legend(loc=0)
plt.title("Blackbody Spectral Radiance of Earth and Sun")
plt.show()

plt.figure(figsize=(7,4))
plt.plot(wlen, p_intensity_sun/np.max(p_intensity_sun), label="Sun ($T=6000$ K)")
plt.plot(wlen, p_intensity_earth/np.max(p_intensity_earth), label="Earth ($T=283$ K)")
plt.axvline(x=4000, color='black', label="Shortwave/Longwave Cutoff")
plt.ylim(0,1.5)
plt.xscale('log')
plt.xlabel("Wavelength [$\mu m$]")
plt.ylabel("Normalized Radiance")
plt.legend(loc=0, ncol=2)
plt.title("Normalized Blackbody Spectral Radiance Given Earth and Sun")
plt.show()
```
