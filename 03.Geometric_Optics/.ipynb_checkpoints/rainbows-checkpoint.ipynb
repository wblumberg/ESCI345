# Geometric Optics Examples

The development of a rainbow by atmospheric radiation and raindrops can be modeled a few different ways.  When the size parameter ($x$) is quite large, often geometric optics can be an efficient way of modeling the behavior of radiation.  For this, you need to know the index of refraction $N$ of the medium you're trying to model.  In the case of rainbows, we're looking at liquid water and air as the two mediums radiation is propagating through. 

Ray tracing uses Snell's law to model the evolution of a ray of light as it travels.

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
index_of_refractions = pd.read_csv('data/visible_refractive_indices.csv')

index_of_refractions
```

```python
plt.plot(index_of_refractions['Wavelength (nm)'], index_of_refractions[' Refractive Index Real'], label=r"$n_r$")
#plt.plot(index_of_refractions['Wavelength (nm)'], index_of_refractions[' Refractive Index Imaginary'], label=r"$n_i$")
plt.yscale('log')
plt.grid()
plt.ylabel("Value")
plt.xlabel('Wavelength (nm)')
plt.legend()
plt.show()
```

```python
impact_parameter = np.arange(0,1.002,0.002)

def viewing_angle(n_water, ip, n_air=1):
    theta = 4 * np.degrees(np.arcsin((n_air/n_water) * ip)) - 2 * np.degrees(np.arcsin(ip))
    return theta

def viewing_angle2(n_water, ip, n_air=1):
    theta = 180 + 2 * np.degrees(np.arcsin(ip)) - 6*np.degrees(np.arcsin((n_air/n_water) * ip))
    return theta

fig = plt.figure()
for i, row in index_of_refractions.iterrows():
    #print(row)
    p_0 = viewing_angle(row[' Refractive Index Real'], impact_parameter)
    p_1 = viewing_angle2(row[' Refractive Index Real'], impact_parameter)

    plt.plot(impact_parameter, p_0, c=row[' Color'].strip())
    plt.plot(impact_parameter, p_1, c=row[' Color'].strip())
plt.grid()
plt.ylim(36,58)
plt.xlim(0.7,1)
plt.ylabel(r"Viewing Angle [deg above horizon]")
plt.xlabel(r"Impact Parameter [unitless]")
plt.show()
```
