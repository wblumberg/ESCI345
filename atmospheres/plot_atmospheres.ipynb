```python
import pandas as pd
import glob
import numpy as np
import proplot as pplt
import pyarts
pplt.use_style('seaborn-poster')
pplt.rc["legend.facecolor"] = "white"
```

```python
atmosphere = ['midlatitude-summer', 'midlatitude-winter', 'subarctic-summer', 'subarctic-winter', 'tropical']
atmospheres = {}
for atmos in atmosphere:
    files = glob.glob(f"{atmos}//*.*.xml")
    data_dict = {}
    for f in files:
        if 'expanded' in f:
            continue
            
        print(f)
        data = pyarts.xml.load(f)
        data_xr = data.to_xarray()

        var = f.split('/')[1].split('.')[1]
        if var != "z" and var != "t":
            data_vals = data_xr
        else:
            data_vals = data_xr
        #print(data_vals.shape, var)
        data_dict[var] = data_vals.to_numpy().squeeze()
        pressure = data_xr.coords["Pressure"].to_numpy().squeeze()
        #print(pressure.shape, "pressure")
    pressure = data_xr.coords["Pressure"].to_numpy().squeeze()
    #print(pressure.shape, "pressure")
    data_dict['p'] = data_xr.coords["Pressure"].to_numpy().squeeze()
    #print(data_dict)
    data_df = pd.DataFrame(data_dict)
    data_df['z'] = data_df['z']#/1000.
    data_df = data_df.set_index('z')
    data_df = data_df.reindex(sorted(data_df.columns), axis=1)
    print(data_df)
    data_df.to_csv(f"{atmos}.csv")
    atmospheres[atmos] = data_df
```

```python
for atmos in atmospheres.keys():
    data = atmospheres[atmos]
    print(data)
    fig, axs = pplt.subplots(ncols=1, nrows=1, figsize=(12,9))
    #axs[0].plot(data["N2"], data.index, label=r"$N_2$")
    axs[0].plot(data["CO"], data.index, label=r"$CO$")
    axs[0].plot(data["O2"], data.index, label=r"$O_2$")
    axs[0].plot(data["H2O"], data.index, label=r"$H_2 O$")
    axs[0].plot(data["CH4"], data.index, label=r"$CH_4$")
    axs[0].plot(data["CO2"], data.index, label=r"$CO_2$")
    axs[0].plot(data["N2O"], data.index, label=r"$N_2 O$")
    axs[0].plot(data["O3"], data.index, label=r"$O_3$")

    axs.format(xlim=(1e-8,100), xscale='log', ylim=(0,100), ylabel="Height [km AGL]", xlabel="Concentration by Volume [%]", grid=True, xformatter='sci')
    fig.legend(loc='bottom', ncols=9)
    fig.format(title=f"{atmos} Atmosphere")
    fig.save(f"{atmos}_gasconc.eps")
```

```python
fig, axs = pplt.subplots(ncols=1, nrows=1, figsize=(10,9))
for atmos_label in atmospheres.keys():
    data = atmospheres[atmos_label]
    axs[0].plot(data["t"].to_numpy(), data.index, label=atmos_label.upper())
axs.format(xreverse=False, ylim=(0,100), ylabel="Height [km AGL]", xlabel="Temperature [K]", grid=True)
fig.legend(loc='right', ncols=1)
fig.format(title="Atmospheres")
fig.save("atmos_thermo.eps")
```
