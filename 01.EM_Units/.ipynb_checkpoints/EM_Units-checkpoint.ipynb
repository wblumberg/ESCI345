### Describing EM Radiation

There are other ways we can modify the x-axis to use different units to describe where we are in the electromagnetic spectrum.  Here are some of the ones we talked about (you can reference your book for the equations).

- Wavelength ($\lambda$)
- Frequency ($f$)
- Wavenumber ($\nu$)
- Photon energy ($E$)

It is sometimes helpful to write functions that can do the conversions:

```python
c = 3e8 # m/s (speed of light)
h = 6.626e-34 # J s (Planck's Constant)

def wl2freq(wavelength):
    # Input: Wavelength in meters
    # Output: Frequency in Hz
    
    frequency = c/wavelength
    return frequency

def wl2wnum(wavelength):
    # Input: Wavelength in meters
    # Output: Wavenumber in meters-1
    
    return 1./wavelength

def freq2energy(frequency):
    # Input: Frequency in Hz
    # Output: Energy in Joules
    
    energy = h * frequency
    return energy

wavelength = data["Wavelength (nm)"]

# Now we store the other variables in the DataFrame
data['Frequency (Hz)'] = wl2freq(wavelength)
data['Wavenumber (m-1)'] = wl2wnum(wavelength)
data['Energy (J)'] = freq2energy(data['Frequency (Hz)'])
```

```python
data
```

You may have noticed that now there are 5 rows.  This is a convenient way to keep track of all of your data - just add it to the DataFrame.  Let's now make three plots.  Each one should have a different x-axis for each different unit type (frequency, wavenumber, and photon energy).

```python
plt.figure(figsize=(9,3))
plt.plot(data["Frequency (Hz)"], data["W*m-2*nm-1"]) # NEW CODE!
plt.xlabel("Frequency (Hz)") # NEW CODE!
plt.ylabel("Solar Irradiance W*m-2*nm-1")
plt.ylim(0,2.5) # NEW CODE!
plt.grid()
plt.tight_layout()
plt.show()

plt.figure(figsize=(9,3))
plt.plot(data["Wavenumber (m-1)"], data["W*m-2*nm-1"]) # NEW CODE!
plt.xlabel("Wavenumber (m-1)") # NEW CODE!
plt.ylabel("Solar Irradiance W*m-2*nm-1")
plt.ylim(0,2.5) 
plt.grid()
plt.tight_layout()
plt.show()

plt.figure(figsize=(9,3))
plt.plot(data["Energy (J)"], data["W*m-2*nm-1"]) # NEW CODE!
plt.xlabel("Photon Energy (J)") # NEW CODE!
plt.ylabel("Solar Irradiance W*m-2*nm-1")
plt.ylim(0,2.5) # NEW CODE!
plt.grid()
plt.tight_layout()
plt.show()
```
