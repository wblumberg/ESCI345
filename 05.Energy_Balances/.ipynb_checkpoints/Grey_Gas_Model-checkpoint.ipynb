# Energy Balance Model - Grey Atmosphere Radiation Model

Below are some import statements to help us set up this model.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
from numpy import cos, deg2rad, log
import climlab
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

### 2-Layer Radiative Balance Model

Here we initialize the model parameters.  We want a model with two atmospheric layers.  We specify the initial temperatures, the longwave absorptivity of the layers, and the number of days for the simulation.

We can include other parameters, but this is really all we need for the time being.

```python
#  Test in a 2-layer atmosphere
col = climlab.GreyRadiationModel(num_lev=2)

# Set up initial state variables (temperatures)
col.Ts[:] = 288. # This is the surface temperature
col.Tatm[:] = np.array([230,275]) # This is the temperature at the two levels of the atmosphere

num_days = 300 # This is how many days we integrate forward.

# Set up initial longwave absorptivity
LW = col.subprocess['LW'] # 
LW.absorptivity = np.array([.59,.59]) # Set up the profile showing the absorptivity of the longwave radiation
col.compute_diagnostics()
```

Here's where we advance the model's state forward and store the diagnostic and state variables we need to understand the model's evolution.

```python
# Set up some arrays to store the model state variables and diagnostics
diagnostics_arr = []
state_arr = []

OLR = [] # outgoing longwave radiation
LW_absorbed_atm = []
SW_up_TOA = []

# Loop over every day and integrate forward the model.
for i in range(0,num_days):
    col.compute_diagnostics()
    diag = climlab.to_xarray(col.diagnostics)
    st = climlab.to_xarray(col.state).copy(deep=True)

    # store some of the diagnostics (climlab did something weird that didn't make the conversion to xarray work perfectly).
    OLR.append(col.OLR[0])
    LW_absorbed_atm.append(col.LW_absorbed_atm[0])
    SW_up_TOA.append(col.SW_up_TOA[0])
    
    diagnostics_arr.append(diag)
    state_arr.append(st)
    
    col.step_forward()

# Create the time dimension
time_dim = pd.RangeIndex(start=0, stop=col.time['days_elapsed'], step=1, name='day')

# Create the xarray data showing all of the timesteps of the diagnostics and state variables.
diagnostics_xr = xr.concat(diagnostics_arr, dim=time_dim, compat="equals")
state_xr = xr.concat(state_arr, dim=time_dim)

# Merge all of the model results and add in the OLR data.
model_results = xr.merge([diagnostics_xr,state_xr])
model_results = model_results.assign({'OLR': (('day',), np.asarray(OLR))})

model_results
```

## Plotting the model outputs over time...

```python
fig, ax = plt.subplots(ncols=2, nrows=1, figsize=(12,4))
plt.axes(ax[0])
plt.plot(model_results.day, model_results['Tatm'], label=["Layer 2", "Layer 1"])
plt.plot(model_results.day, model_results['Ts'], label="Surface Temperature")
plt.legend(fontsize=10, loc='upper center', bbox_to_anchor=(0.5, 1.1),
          ncol=3, fancybox=True, shadow=True)
plt.xlabel("Simulation Time [days]")
plt.ylabel("Temperature [Kelvin]")
plt.grid()

plt.axes(ax[1])
plt.plot(model_results.day, model_results['ASR'] - model_results['OLR'])
plt.ylabel("$ASR-OLR$ [$W\ m^{-2}$]")
plt.title("Radiation Balance")
plt.xlabel("Simulation Time [days]")
plt.grid()
plt.tight_layout()
plt.show()
```

```python
model_results['LW_emission'][:,0].plot(label="Layer 1 - Upwelling LW Flux")
model_results['LW_emission'][:,1].plot(label="Layer 2 - Upwelling LW Flux")
model_results['LW_up_sfc'].T.plot(label="Surface - Upwelling LW Flux")
model_results['SW_down_TOA'].T.plot(label="TOA SW Downwelling")

plt.grid()
plt.ylabel("Flux [$W\ m^{-2}$]")
plt.title("Radiative Balance Components")
plt.xlabel("Simulation Time [days]")
plt.legend()
plt.show()
plt.plot(model_results.day, model_results['ASR'] - model_results['OLR'], 'b')
plt.ylabel("$ASR-OLR$ [$W\ m^{-2}$]", color='b')
plt.xlabel("Simulation Time [days]")
plt.twinx()
plt.plot(model_results.day, model_results['Ts'], 'r-')
plt.ylabel("Surface Temperature [$K$]", color='r')
plt.grid()
plt.show()
```

## 10-layer Radiative Balance Model

```python
#  Test in a 2-layer atmosphere
col = climlab.GreyRadiationModel(num_lev=10)

# Set up initial state variables (temperatures)
col.Ts[:] = 288. # This is the surface temperature
col.Tatm[:] = np.linspace(230,286,10) # This is the temperature at the two levels of the atmosphere

num_days = 250 # This is how many days we integrate forward.

# Set up initial longwave absorptivity
LW = col.subprocess['LW'] # 
LW.absorptivity = 0.18 * np.ones(10) # Set up the profile showing the absorptivity of the longwave radiation
col.compute_diagnostics()

# Set up some arrays to store the model state variables and diagnostics
diagnostics_arr = []
state_arr = []

OLR = [] # outgoing longwave radiation
LW_absorbed_atm = []
SW_up_TOA = []

# Loop over every day and integrate forward the model.
for i in range(0,num_days):
    col.compute_diagnostics()
    diag = climlab.to_xarray(col.diagnostics)
    st = climlab.to_xarray(col.state).copy(deep=True)

    # store some of the diagnostics (climlab did something weird that didn't make the conversion to xarray work perfectly).
    OLR.append(col.OLR[0])
    LW_absorbed_atm.append(col.LW_absorbed_atm[0])
    SW_up_TOA.append(col.SW_up_TOA[0])
    
    diagnostics_arr.append(diag)
    state_arr.append(st)
    
    col.step_forward()

# Create the time dimension
time_dim = pd.RangeIndex(start=0, stop=col.time['days_elapsed'], step=1, name='day')

# Create the xarray data showing all of the timesteps of the diagnostics and state variables.
diagnostics_xr = xr.concat(diagnostics_arr, dim=time_dim, compat="equals")
state_xr = xr.concat(state_arr, dim=time_dim)

# Merge all of the model results and add in the OLR data.
model_results = xr.merge([diagnostics_xr,state_xr])
model_results = model_results.assign({'OLR': (('day',), np.asarray(OLR))})
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10,4))
model_results['LW_emission'].T.plot(ax=axs[0])
axs[0].set_ylim(1000,0)
model_results['Tatm'].T.plot(ax=axs[1])
axs[1].set_ylim(1000,0)
plt.tight_layout()
plt.show()

plt.plot(model_results.day, model_results['ASR'] - model_results['OLR'], 'b')
plt.ylabel("$ASR-OLR$ [$W\ m^{-2}$]", color='b')
plt.twinx()
plt.plot(model_results.day, model_results['Ts'], 'r-')
plt.ylabel("Surface Temperature [$K$]", color='r')
plt.grid()
plt.xlabel("Simulation Time [days]")
plt.show()
```
