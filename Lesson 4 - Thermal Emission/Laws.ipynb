```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
c = 2.99792458e8 # m/s
k_b = 1.380649e-23 # J K-1
h = 6.62607015e-34 # J s

def rayleigh_jeans(wavelength, temperature):
    # UV catastrophy
    intensity = (2. * c * k_b * temperature)/np.power(wavelength,4)
    return intensity

def wien_law(wavelength, temperature):
    multi = (2. * h * c**2)/np.power(wavelength,5.)
    exp = np.exp( - (h * c)/(wavelength * k_b * temperature))
    return multi * exp

def planck(wavelength, temperature):
    multi = (2. * h * c**2)/np.power(wavelength,5.)
    a = np.exp((h * c)/(wavelength * k_b * temperature))
    intensity = multi * (1./(a - 1.))

    return intensity
```

```python
wlen = np.arange(.01,3,.01) / 1e6 # to meters
rj_intensity = rayleigh_jeans(wlen, 6000.) * 1e-9
wien_intensity = wien_law(wlen, 6000.) * 1e-9
p_intensity = planck(wlen, 6000.) * 1e-9
```

```python
sns.set_theme()
plt.figure(figsize=(7,4))
plt.plot(wlen * 1e6, rj_intensity, label="Rayleigh-Jeans Law")
plt.plot(wlen * 1e6, wien_intensity, label="Wien's Law")
plt.plot(wlen * 1e6, p_intensity, label="Planck's Law")
#plt.yscale('log')
plt.ylim(0,5e13 * 1e-9)
plt.xlabel("Wavelength [$\mu m$]")
plt.ylabel("Spectral Radiance [$W\ m^{-2}\ sr^{-1}\ nm^{-1}$]")
plt.legend(loc=0)
plt.title("Spectral Radiance at $T$ = 6000 K")
```
